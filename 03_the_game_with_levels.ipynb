```python
import random

# print welcome message
def print_welcome_message():
    print("Welcome to the number guessing game!")
    print("Try to guess the secret number within the guess limit.")

# print difficulty options and return the range based on user choice
def choose_difficulty():
    print("\nChoose a difficulty level:")
    print("1. Easy   (1-50,  10 guesses)")
    print("2. Medium (1-100, 10 guesses)")
    print("3. Hard   (1-500, 10 guesses)")

# while loop to ensure valid input for difficulty choice
    while True:
        choice = input("Enter 1, 2, or 3: ")
        if choice == "1":
            return 1, 50
        elif choice == "2":
            return 1, 100
        elif choice == "3":
            return 1, 500
        else:
            print("Please enter 1, 2, or 3.")

def get_valid_guess(low, high, previous_guesses):
    while True:
        guess_input = input(f"What is your guess? (The number is between {low} and {high} inclusive): ")

        try:
            guess = int(guess_input)
        except ValueError:
            print("That is not correct number. Please enter a another number.")
            continue

        if guess < low or guess > high:
            print(f"Please guess a number between {low} and {high}.")
            continue

        if guess in previous_guesses:
            print(f"You already guessed {guess}. Try a different number.")
            continue

        return guess

print_welcome_message()
low, high = choose_difficulty()

secret_number = random.randint(low, high)
num_guesses_remaining = 10
is_still_guessing = True
previous_guesses = []

print(f"\nThe secret number is between {low} and {high} inclusive. Good luck!\n")

# while loop to allow the user to keep guessing until they run out of guesses or guess the secret number
while num_guesses_remaining > 0 and is_still_guessing:

    guess = get_valid_guess(low, high, previous_guesses)
    previous_guesses.append(guess)

    if guess > secret_number:
        print("Sorry, you guessed too high!")
    elif guess < secret_number:
        print("Its actually bigger than that...")
    else:
        print("That is right!! Well done.")
        is_still_guessing = False

    num_guesses_remaining -= 1

    if is_still_guessing and num_guesses_remaining > 0:
        print(f"You have: {num_guesses_remaining} guesses left.")
        print(f"Guesses so far: {previous_guesses}\n")

# if the user runs out of guesses, print a message and reveal the secret number
if is_still_guessing and num_guesses_remaining == 0:
    print("Sorry, you're out of guesses :( ")
    print(f"The secret number was: {secret_number}")
```
