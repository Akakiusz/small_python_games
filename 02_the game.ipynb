```python
# import random
import random
secret_number = random.randint(1, 100,)
num_guesses_remaining = 10
is_still_guessing = True

def print_welcome_message():
    print("Welcome to the number guessing game!")
    print("You have 10 guesses to guess the secret number.")
    print("The secret number is between 1 and 100 inclusive.")

print_welcome_message()

# WHILE LOOP
while num_guesses_remaining > 0 and is_still_guessing:

    guess = int(input("What is your guess? (The number is between 1 and 100 inclusive)"))

    if guess > secret_number:
        print("Sorry, you guessed too high!")
    elif guess < secret_number:
        print("Its actually bigger than that...")
    else:
        print("That is right!! Well done.")
        is_still_guessing = False
        
    # num_guesses_remaining = num_guesses_remaining - 1
    num_guesses_remaining -= 1
    print(f"You have: {num_guesses_remaining} guesses left.")

if is_still_guessing:
    print("Sorry, you're out of guesses :( ")
    print(f"The secret number was: {secret_number}")
```
